```python
# 마우스 핸들링

import pyautogui as p

#마우스 좌표 구하기
position = p.position()
print(position.x,position.y)
```

```python
743,11,770,27
p.screenshot("file.png",region=(743,11,30,25))
```

```python
file_locate = p.locateOnScreen("file.png",confidence=0.9)
print(file_locate)
p.click(file_locate)
```

```python
# confidence 신뢰도 0.8~0.9
checkbox = p.locateAllOnScreen("checkbox.png",confidence=0.9)
print(checkbox)
p.click(checkbox)
```

```python
for w in p.getAllWindows():
    print(w)
```

```python
for w in p.getActiveWindowTitle():
    print(w)
```

```python
for w in p.getWindowsWithTitle("file.png"):
    print(w)
```

```python
w = p.getWindowsWithTitle("file.png")[0]

print(w)

if not w.isActive:
    w.activate()
```

```python
if not w.isMaximized():
    w.maximize()

if not w.isMinimized():
    w.minimize()
```

```python
# 창 원래대로 
w.restore()
```

```python
# 창 종료
w.close()
```
