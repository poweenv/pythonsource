```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# warnings : 함수를 현재는 지원하나 나중에 삭제 시 경고 메세지가 발생
import warnings
warnings.simplefilter("ignore")

sns.set_theme()
sns.set(rc={"figure.figsize":(12,8)})

plt.rcParams["font.family"]="Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
movie_title=["크루엘라","극장판 귀멸의 칼날 : 무한열차편","학교 가는 길"]
audience = [664308,2099131,20067]

df = pd.DataFrame(
    {
        "영화제목" : movie_title,
        "누적관객" : audience
    }
)
df
```

```python
sns.barplot(data=df,x="영화제목",y="누적관객")
```

```python
sns.barplot(data=df,x="영화제목",y="누적관객",order=df.sort_values("누적관객").영화제목)
```

```python
sns.barplot(data=df,x="영화제목",y="누적관객",order=df.sort_values("누적관객",ascending=False).영화제목)
```

```python
g=sns.barplot(data=df,x="영화제목",y="누적관객",order=df.sort_values("누적관객",ascending=False).영화제목)

label_format="{:,.0f}"

ylabels =[label_format.format(i) + "만 명" for i in g.get_yticks() / 10000]

g.set_yticklabels(ylabels)
```
