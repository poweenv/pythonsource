```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.rcParams["font.family"]="Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
# warnings : 함수를 현재는 지원하나 나중에 삭제 시 경고 메세지가 발생
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.DataFrame(np.random.randn(1000,3),columns=list('ABC'))
df = df.cumsum()
df
```

```python
df.plot()
```

```python
g = df.plot(legend=False)
```

```python
g = df.plot(legend=False)
```

```python
g=df.plot(y="B")
```

```python
# 보조출 설정

df["A"].plot()

df["B"].plot(secondary_y=True,style="r")
```

```python
# 보조축 + 레이블 지정

#ax = df.plot(secondary_y=["A","B"])  범례에 right 표시
ax = df.plot(secondary_y=["A","B"],mark_right=False)
ax.set_ylabel("C_Scale")
ax.right_ax.set_ylabel("AB_Scale")
plt.show()
```

```python
# 컬럼별 선 차트 따로 그리기
g=df.plot(subplots=True)
```

```python
np.random.seed(1000)

df = pd.DataFrame(np.random.randn(1000,4),columns=list('ABCD'),index=pd.date_range('1/1/2021',periods=1000))

df
```

```python
# 2021-01-06 날짜로 bar chart 작성
# iloc : 숫자, loc : 레이블명
df.iloc[5].plot(kind="bar")
```

```python
df2 = pd.DataFrame(np.random.rand(50,4),columns=list('abcd'))
```

```python
df2.plot.scatter(x="a",y="b")
```

```python
# s : 점 크기 설정
df2.plot.scatter(x="a",y="b",s=50)
```

```python
ax=df2.plot.scatter(x="a",y="b",color="DarkBlue",label="Group 1")
df2.plot.scatter(x="c",y="d",color="DarkGreen",label="Group 2", ax=ax)
```
