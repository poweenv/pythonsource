```python

# from selenium.webdriver.common.by import By
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC
# from selenium.webdriver.common.keys import Keys
# import time
# from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup

# 다나와 노트북 선택
url = "https://prod.danawa.com/list/?cate=112758&15main_11_02"

driver = set_chrome_driver()
driver.get(url)
# 제조사 22개 클릭 =>  apple 클릭

#driver.implicitly_wait(2)
time.sleep(4)

try:
    WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,'//*[@id="dlMaker_simple"]/dd/div[2]/button[1]'))).click()
    WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,'//*[@id="selectMaker_simple_priceCompare_A"]/li[16]/label'))).click()

    # 애플 제품 로딩
    time.sleep(3)

    cur_page, target_crawl_num = 1,4
    idx=1
    while cur_page <= target_crawl_num:
    #페이지 나누기 변수

        soup = BeautifulSoup(driver.page_source,"lxml")

        # 상품목록 리스트 모두 가져오기 li:not를 하면 해당되는 것들은 포함하지 않음
        products=soup.select("div.main_prodlist.main_prodlist_list > ul > li:not(.prod_ad_item):not(.product-pot)")
        #print(products)
        for product in products:
            #    번호, 제품명, 가격, 이미지주소
            product_name=product.select_one(".prod_name > a").get_text().strip()
            product_price=product.select_one("p.price_sect strong").text.strip()
            img=product.select_one(".thumb_image img")

            if img.get("data-original"):
                product_img = img.get("data-original")
            else:
                product_img = img.get("src")

            if "http:" not in product_img:
                product_img = "http:"+product_img

            print(idx,product_name,product_price,product_img)
            idx+=1
        
        # 현재 페이지 번호 변경
        cur_page += 1

        # 다음 페이지 번호 클릭
        # div.number_wrap > a:nth-child(cur_page)
        WebDriverWait(driver,2).until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR,"div.number_wrap > a:nth-child({})".format(cur_page))
            )
        ).click()

        # 기존의 soup 삭제
        del soup
        time.sleep(3)


except TimeoutException as e:
    print(e)



```

```python
#상품 정보 엑셀 저장 + img 저장

import requests
from fake_useragent import UserAgent

from io import BytesIO
from openpyxl.drawing.image import Image
from openpyxl import Workbook

# 엑셀 작업
wb = Workbook()
ws = wb.active

ws.column_dimensions["A"].width = 60
ws.column_dimensions["B"].width = 20

# 제목행 추가
ws.append(["제품명","가격","이미지"])

# 다나와 노트북 선택
url = "https://prod.danawa.com/list/?cate=112758&15main_11_02"

driver = set_chrome_driver()
driver.get(url)
# 제조사 22개 클릭 =>  apple 클릭

#driver.implicitly_wait(2)
time.sleep(4)

try:
    WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,'//*[@id="dlMaker_simple"]/dd/div[2]/button[1]'))).click()
    WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,'//*[@id="selectMaker_simple_priceCompare_A"]/li[16]/label'))).click()

    # 애플 제품 로딩
    time.sleep(3)

    cur_page, target_crawl_num = 1,4
    # 전체 상품 번호
    idx=1

    #엑셀 행 번호
    ins_cnt =2

    while cur_page <= target_crawl_num:
    #페이지 나누기 변수

        soup = BeautifulSoup(driver.page_source,"lxml")

        # 상품목록 리스트 모두 가져오기 li:not를 하면 해당되는 것들은 포함하지 않음
        products=soup.select("div.main_prodlist.main_prodlist_list > ul > li:not(.prod_ad_item):not(.product-pot)")
        #print(products)
        for product in products:
            #    번호, 제품명, 가격, 이미지주소
            product_name=product.select_one(".prod_name > a").get_text().strip()
            product_price=product.select_one("p.price_sect strong").text.strip()
            img=product.select_one(".thumb_image img")

            if img.get("data-original"):
                product_img = img.get("data-original")
            else:
                product_img = img.get("src")

            if "http:" not in product_img:
                product_img = "http:"+product_img

            #print(idx,product_name,product_price,product_img)

            # 이미지 요청
            res = requests.get(product_img,headers={"user-agent" : UserAgent().chrome})
            # res.content : byte 코드로 결과  가져오기
            # res.text : text 결과 가져오기
            save_img = BytesIO(res.content)
            # 상품 하나에 대한 정보 저장
            ws.append([product_name,product_price])

            image = Image(save_img)
            image.width=30
            image.height=20
            ws.add_image(image,"C"+str(ins_cnt))
            idx+=1
            ins_cnt +=1
        # 현재 페이지 번호 변경
        cur_page += 1

        # 다음 페이지 번호 클릭
        # div.number_wrap > a:nth-child(cur_page)
        WebDriverWait(driver,2).until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR,"div.number_wrap > a:nth-child({})".format(cur_page))
            )
        ).click()

        # 기존의 soup 삭제
        del soup
        time.sleep(3)

    
except TimeoutException as e:
    print(e)

wb.save("./apple.xlsx")


wb.close()
```
